## Data Cleaning

```python
nycdata = data.drop(labels = ['Incident Address', 'Street Name', 'Cross Street 1',
                   'Intersection Street 1', 'Intersection Street 2', 'School or Citywide Complaint', 
                    'Vehicle Type', 'Taxi Company Borough', 
                    'Taxi Pick Up Location', 'Bridge Highway Name', 'Bridge Highway Direction', 
                    'Road Ramp', 'Bridge Highway Segment', 
                    'Garage Lot Name', 'Ferry Direction', 'Ferry Terminal Name', 'Landmark'], axis = 1)
```

```python
nycdata.dropna(inplace = True)
```

```python
for i in nycdata.iloc[:,22:33].columns:
    print(nycdata[i].value_counts())
```

```python
nycdata.drop(nycdata.iloc[:,22:33], axis = 1, inplace = True)
```

```python
nycdata['Park Facility Name'].value_counts()
```

```python
nycdata.drop('Park Facility Name', axis = 1, inplace = True)
```

```python
nycdata.shape
```

```python
nycdata['Location Type'].nunique()
```

```python
nycdata.head()
```

```python
nycdata['Created Date'] = pd.to_datetime(nycdata['Created Date'])
nycdata['Closed Date'] = pd.to_datetime(nycdata['Closed Date'])
nycdata['Due Date'] = pd.to_datetime(nycdata['Due Date'])
nycdata['Resolution Action Updated Date'] = pd.to_datetime(nycdata['Resolution Action Updated Date'])
```

## Task 1 

```python
nycdata['Request_closing_time'] = nycdata['Closed Date'] - nycdata['Created Date']
```

```python
nycdata['Request_closing_time_secs'] = nycdata['Request_closing_time'].dt.seconds
```

```python
nycdata['Request_closing_time_secs'].head()
```

```python
nycdata['TPD'] = nycdata['Resolution Action Updated Date']-nycdata['Due Date']
```

*TPD Stands for Time past due Date

```python
nycdata['TPD in secs'] = nycdata['TPD'].dt.seconds
```

```python
nycdata['TPD in secs'].head()
```

#### Insight 4 with Graph

```python
insight4 = pd.DataFrame(nycdata.groupby(['City'])['Request_closing_time_secs', 'TPD in secs'].mean().sort_values(by = 'Request_closing_time_secs',ascending = False))
insight4.reset_index(inplace = True)
insight4
```

*TPD Stands for time past due date

```python
fig = plt.figure(figsize=(15,8))
ax1 = plt.subplot()
l1, = ax1.plot(insight4['City'],insight4['Request_closing_time_secs'], color='red')
plt.xlabel('City')
plt.xticks(rotation = 90)
ax2 = ax1.twinx()
l2, = ax2.plot(insight4['City'],insight4['TPD in secs'], color='green')
a,b = 0,80000
ax1.set_ylim(a,b)
ax2.set_ylim(a,b)
plt.legend([l1, l2], ["Average Request closing time secs", "Average TPD in secs"])
plt.show();
```

- Floral Park was 6th lowest in no of complaints but taking highest avg time in request closing.
- Brooklyn has the highest no of complaints but stands 35th(out of 49) when compared avg time taken to close request.
- Howard Beach has taken highest avg time after Due date follwed by Arverne.

#### Insight 5 with Graph

```python
x = pd.DataFrame(nycdata.groupby(['Agency Name', 'Complaint Type'])['Complaint Type'].count().sort_values(ascending = False))
x.rename(columns = {'Complaint Type': "Complaint Count"}, inplace = True)
x.reset_index(inplace = True)
y = pd.DataFrame(nycdata.groupby(['Agency Name', 'Complaint Type'])['Request_closing_time_secs', 'TPD in secs'].mean().sort_values(by = 'Request_closing_time_secs',ascending = False))
y.reset_index(inplace = True)

insight5 = x.merge(y, on = ['Agency Name', 'Complaint Type'])
insight5
```

- Disoderly youth complaint type has the highest average time past due date to close.
- Graffiti took highest avg time to close request.

```python
from scipy import stats
from scipy.stats import ttest_ind, pearsonr
```

```python
stats.ttest_ind

complaint = []
avg_time = []

complaints = nycdata['Complaint Type'].unique()

for i in complaints:
    m = nycdata[nycdata['Complaint Type'] == i]['Request_closing_time_secs'].mean()
    complaint.append(i)
    avg_time.append(m)
    
comp = pd.DataFrame({"Complaint Type": complaint, "Average Time": avg_time})
comp
```

```python
for i in complaints:
    x = np.argwhere(complaints == i)
    x = int(x)
    y = np.delete(complaints, x)
    for j in y:
        
        set1 = nycdata[nycdata['Complaint Type'] == i]['Request_closing_time_secs'].dropna()
        set2 = nycdata[nycdata['Complaint Type'] == j]['Request_closing_time_secs'].dropna()
        
        statistic, p_value = ttest_ind(set1, set2)
        
        print(round(p_value,3), "-----> p value for similar average for", i, ",", j, "\n\n")
```
